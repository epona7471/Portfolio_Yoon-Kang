```python
import wave
import os

pcm_file = '/content/drive/MyDrive/korean_english/GSF04623LWJ0/SW201105ETRNGSF04623LWJ0003.pcm'
save_file = '/content/drive/MyDrive/korean_english/temp.wav'

channels=1
bits=16
sample_rate=16000

pcmf = open(pcm_file, 'rb')
pcmdata = pcmf.read()
pcmf.close()


wavfile = wave.open(save_file, 'wb')

wavfile.setnchannels(channels)
wavfile.setsampwidth(bits // 8)
wavfile.setframerate(sample_rate)

wavfile.writeframes(pcmdata)
wavfile.close()

```
